# TENSOR INPUT


```python
# y = x.w + b

w = tf.Variable(tf.random.normal((3, 2)), name = 'w')
print(w)

b = tf.Variable(tf.zeros((1,2), dtype=tf.float32), name = 'b')
# b = tf.Variable(tf.zeros(2, dtype=tf.float32), name = 'b')  also works due to tf broadcasting
print(b)

x = tf.constant([[1.0,2.0,3.0]])
print(x)
```

```python
with tf.GradientTape(persistent=True) as tape:
    y = x @ w + b  # @ is the sign for dot product
    loss = tf.reduce_mean(y**2)
```

```python
[dl_dw, dl_db] = tape.gradient(loss, [w,b])
print(dl_dw)
print(dl_db)
```

# Dictionary Input



```python
w = tf.Variable(tf.random.normal((3,2)), name = 'w')

x = tf.constant([[1.0,2.0,3.0]])

b = tf.Variable(tf.zeros((1,2)), name = 'b')

with tf.GradientTape(persistent=True) as tape:
    y = x @ w + b
    loss = tf.reduce_mean(y**2)

my_vars = {
    'w': w,
    'b': b
}

grad = tape.gradient(loss, my_vars)
print(grad['w'])
print(grad['b'])
```

# Gradiens w.r.t a model

```python
layer = tf.keras.layers.Dense(2, activation='relu')

x = tf.constant([[1.0,2.0,3.0]])

# Record
with tf.GradientTape() as tape:
    y = layer(x)
    loss = tf.reduce_mean(y**2)

grad = tape.gradient(loss, layer.trainable_variables)

print (grad)
```
